```python
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv', low_memory = False)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.head()
```

```python
df.reset_index(inplace = True, drop = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Es posible convertir la fecha a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], format = "mixed", dayfirst = True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Ver los valores para el estatus
df['Estatus'].unique()
```

```python
#Corregir error ortografico del estatus
map_estatus = {
    "revisado": "REVISADO",
    "verificado": "VERIFICADO"
}
df['Estatus'] = df['Estatus'].map(map_estatus)
```

```python
df['Estatus'].unique()
```

```python
#Separar las fechas en dia, mes, año y de ser posible, dia de la semana
df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month
df['Dia'] = df['Fecha'].dt.day

df.head(5)
```

```python
import calendar
```

```python
df['Dia Semana'] = df['Fecha'].dt.day_name()
df.head(5)
```

```python
map_semana = {
    "Monday" : "LUNES",
    "Tuesday" : "MARTES",
    "Wednesday" : "MIERCOLES",
    "Thursday" : "JUEVES",
    "Friday" : "VIERNES",
    "Saturday" : "SABADO",
    "Sunday" : "DOMINGO"
}
```

```python
df['Dia Semana'] = df['Dia Semana'].map(map_semana)
df.head(5)
```

```python
#Eliminar columna del UTC
df = df.drop('Fecha UTC', axis = 1)
```

```python
df = df.drop('Hora UTC', axis = 1)
```

```python
#Separar las referencias de localizacion en Localidad, Estado, Direccion
df['Estado'] = df['Referencia de localizacion'].str.split(",").str[1].str.strip()
df.head()
```

```python
df['Localidad'] = df['Referencia de localizacion'].str.extract(r'de\s+(.*),')
df.head()
```

```python
df['Direccion Cardinal'] = df['Referencia de localizacion'].str.extract(r'\b(AL|DEL|AL)?\s*(NORTE|SUR|ESTE|OESTE|SUROESTE|NOROESTE|SURESTE|NORESTE)\b', expand=False)[1]
df.head()
```

```python
df['Distancia KM'] = df['Referencia de localizacion'].str.extract(r'(\d+)\s*km')
df.head()
```

```python
df = df.drop('Referencia de localizacion', axis = 1)
```

```python
df.head()
```

```python
df['Hora'] = pd.to_datetime(df['Hora'])
df.head()
```

```python
df['Segundo'] = df['Hora'].dt.second
df['Minuto'] = df['Hora'].dt.minute
df['Hora'] = df['Hora'].dt.hour
df.head()
```

```python
orden = ['Fecha', 'Dia', 'Mes', 'Año', 'Dia Semana', 'Hora', 'Minuto', 'Segundo','Estado','Localidad','Distancia KM','Direccion Cardinal','Magnitud','Profundidad','Latitud','Longitud','Estatus']
```

```python
df = df[orden]
```

```python
df.head()
```

```python
df.to_csv('data/processed_data.csv', index=False)
```
